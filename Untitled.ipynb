```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
#getting file 
data = pd.read_excel('Global-Oil-and-Gas-Extraction-Tracker-Jan-2022.xlsx')
```

```python
df = pd.DataFrame(data)
```

```python
oilgas_df= df[df['Status'] == 'in development'].copy()
```

```python
#Calling my dataset
oilgas_df2 = oilgas_df[oilgas_df['Location accuracy'] == "exact"].copy()
```

```python
oilgas_df2.sample(5)
```

```python
# list of desired column names
desired_columns = ['Unit name','Fuel type','Country','Latitude','Longitude','Status','Status year','Discovery year','Production start year','Operator','Parent']
oilgas_df2[desired_columns].sample(5)
```

```python
oilgas_trimmed = oilgas_df2[desired_columns].copy()
```

```python
oilgas_trimmed.sample(5)
```

```python
oilgas_trimmed['geometry']=''
```

```python
oilgas_trimmed.head(2)
```

```python
oilgas_gdf = gpd.GeoDataFrame(
    oilgas_trimmed, geometry=gpd.points_from_xy(oilgas_trimmed.Longitude, oilgas_trimmed.Latitude))
```

```python
oilgas_gdf.head(2)
```

```python
oilgas_gdf.plot()
```

```python
oilgas_gdf.geometry
```

```python
import folium
```

```python
# default folium map
m = folium.Map()
m
```

```python
for index, row in oilgas_gdf.iterrows():
    print(row.Latitude, row.Longitude)
```

```python
for index, row in oilgas_gdf.iterrows():
    folium.Marker([row.Latitude, row.Longitude]).add_to (m)
        
```
