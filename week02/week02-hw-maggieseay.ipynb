```python
#only want to keep ones with exact locations, so setting up a new df that only includes rows with exact locations
oilgas_loc = oilgas_df[oilgas_df['Location accuracy'] == "exact"].copy()
```

```python
#setting up list of desired column names in oilgas_loc df
desired_columns = ['Unit name','Fuel type','Country','Latitude','Longitude','Status','Status year','Discovery year','Production start year','Operator','Parent']
oilgas_loc[desired_columns].sample(2)
```

```python
#making a new df only with desired columns and exact locations
oilgas_trimmed = oilgas_loc[desired_columns].copy()
oilgas_trimmed.sample(2)
```

```python
# got this from the internet, but adding in geometry part so this can be mapped
oilgas_gdf = gpd.GeoDataFrame(oilgas_trimmed, geometry=gpd.points_from_xy(oilgas_trimmed.Longitude, oilgas_trimmed.Latitude))
```

```python
#seeing what the new column looks like
oilgas_gdf.head(2)
```

```python
oilgas_gdf.geometry
```

```python
#creating df that can be mapped of all 'in development' oil/gas projects
oilgas_dev= oilgas_gdf[oilgas_gdf['Status'] == 'in development'].copy()
```

```python
#sampling the 'in development' df
oilgas_dev.sample(3)
```

```python
import folium
```

```python
# default folium map
m = folium.Map()
m
```

```python
#get lat + long for oil+gas in dev
for index, row in oilgas_dev.iterrows():
    print(row.Latitude, row.Longitude)
```

```python
#add to map!
for index, row in oilgas_dev.iterrows():
    folium.Marker([row.Latitude, row.Longitude]).add_to (m)
```
