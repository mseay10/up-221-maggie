```python
#changing my geo_codes to strings
df = pd.read_csv(
    'censusdata_clean.csv',
    dtype=
    {
        'Geo_FIPS':str,
        'Geo_STATE':str,
        'Geo_COUNTY': str
    }
)
```

```python
#aha but still missing leading zero
df.Geo_FIPS.head()
```

```python
df.Geo_STATE.head()
```

```python
#I coundnt figure out why the conversion of Geodata
#did not generate a zero, so I found this fix on the internet.
#Basically says to add zeros until the varibale in the column has 
#(that) many characters
df["Geo_STATE"] = df["Geo_STATE"].str.zfill(2)
df.head()
```

```python
df["Geo_FIPS"] = df["Geo_FIPS"].str.zfill(11)
df.head()
```

```python
df["Geo_COUNTY"] = df["Geo_COUNTY"].str.zfill(3)
df.head()
```

```python
#importing geopandas
import geopandas as gpd
```

```python
#read geojson file I found on the internet!
tracts=gpd.read_file('https://raw.githubusercontent.com/arcee123/GIS_GEOJSON_CENSUS_TRACTS/master/06.geojson')
```

```python
tracts.head()
```

```python
tracts.plot()
```

```python
tracts.info(verbose=True, show_counts=True)
```

```python
tracts = tracts[['GEOID','geometry']]
```

```python
tracts.head()
```

```python
#rename GEOID to prepare for merge
tracts.columns = tracts.columns.str.replace('GEOID', 'Geo_FIPS')
```

```python
#check new column name
tracts.head()
```

```python
#merge!!
salton_data=tracts.merge(df,on="Geo_FIPS")
```

```python
salton_data.head(3)
```

```python
#greating graphs for race/ethnicity
salton_data.plot(column='White Pop',
                 scheme='NaturalBreaks',
                 legend='True'
                )
```
